## `GDAL` and `OGR`

The <b>Geospatial Data Abstraction Library (GDAL)</b> is a computer software library for reading and writing raster and vector geospatial data formats, and is released under the permissive X/MIT style free software license by the Open Source Geospatial Foundation.  https://en.wikipedia.org/wiki/GDAL

`GDAL` deals with raster data manipulation

`OGR` deals with vector data manipulation

GDAL is written in C, C++, Python , and has wrappers for Python, Ruby, R, etc. GDAL is a base library for many industry standard softwares for analaysis, processing, converting data such as 
- ArcGIS
- ERDAS 
- QGIS
- GRASS
- MapServer, etc. 

> Installing GDAL is just has hard as how it makes our life simple


### Installing `GDAL` on windows 
Check whether `GDAL` is already installed on your machine, by opening command line and typing `gdalinfo`, if you see something like this 


you already have GDAL installed, if you see an error saying 
> gdalinfo is not recognized as an internal or external command

then follow along.

Head over to QGIS Downloads page and look for OSGeo4W Network installer. This is the easiest way to install packages in Windows. Download the setup 
<img src='../data/screenshots/osgeo4.png' > 

Once downloaded, open setup and look for <b> osgeo4w-setup.exe</b>, click on <b> advance install</b> , click next till we see  <b> select packages </b>  page, here you can seach for commandline_utilites and search for GDAL, and click next , and it will install gdal in your machine.

Once the installation is done, go to <b>Environment Variables</b>, add Path of `OSGeo4W/bin` and save , now you can open new command prompt and check if GDAL is installed



### Installing `GDAL` on MacOS 

In MacOS, We can install GDAL using `homebrew` 

paste following command in your terminal 

> /bin/bash -c "$(curl -fsSL https://raw.githubusercontent.com/Homebrew/install/HEAD/install.sh)"

Once the brew is installed, confirm it by typing `brew` in terminal.
After this install GDAL 
> brew install gdal

### Installing `GDAL` on ubuntu 

We can follow this https://mothergeo-py.readthedocs.io/en/latest/development/how-to/gdal-ubuntu-pkg.html

```
    sudo add-apt-repository ppa:ubuntugis/ppa
    sudo apt-get update
    sudo apt-get install gdal-bin
    sudo apt-get install libgdal-dev
    export CPLUS_INCLUDE_PATH=/usr/include/gdal
    export C_INCLUDE_PATH=/usr/include/gdal
```
To verify the installation , type following command 

```
    ogrinfo --version
```

To install GDAL in Python
```
   pip install GDAL
```

There are tons of python packages available in geospatial realm, some focuses on solving specific problems, while some are used for general purpose.
